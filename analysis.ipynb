```python
import pandas as pd
```

```python
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
df = pd.read_csv('data/cpshouseholdincome.csv',header=1,thousands=',',na_values='.')

df.rename(columns = {
    'Unnamed: 0':'Race',
    'Unnamed: 1':'Hispanic origin',
    'Unnamed: 2':'Totals',
    '$1 to $2,499\r\nor loss':'$1 to $2,499 or loss',
    '$1 to $2,499\nor loss':'$1 to $2,499 or loss',
    '$100,000 and\r\nover':'$100,000 and over',
    '$100,000 and\nover':'$100,000 and over'
}, inplace = True)

df['Race'] = df['Race'].fillna(method='ffill')

df = df[~df['Totals'].isna()]

df['Hispanic origin'] = df['Hispanic origin'].replace({'Origin': 'Non Hispanic'})
df['Race'] = df['Race'].replace({'Race': 'White alone'})
df.loc[df['Hispanic origin'].isna(), ['Hispanic origin']] = 'Totals'

df
```

```python
brackets = {}

# brackets['No Income'] = [
#    'No Income'
# ]
brackets['Less than $10,000'] = [
    '$1 to $2,499 or loss',
    'No Income',
    '$2,500 to $4,999',
    '$5,000 to $7,499',
    '$7,500 to $9,999'
]
brackets['$10,000 to $20,000'] = [
    '$10,000 to $12,499',
    '$12,500 to $14,999',
    '$15,000 to $17,499',
    '$17,500 to $19,999'
]
brackets['$20,000 to $30,000'] = [
    '$20,000 to $22,499',
    '$22,500 to $24,999',
    '$25,000 to $27,499',
    '$27,500 to $29,999'
]
brackets['$30,000 to $40,000'] = [
    '$30,000 to $32,499',
    '$32,500 to $34,999',
    '$35,000 to $37,499',
    '$37,500 to $39,999'
]
brackets['$40,000 to $50,000'] = [
    '$40,000 to $42,499',
    '$42,500 to $44,999',
    '$45,000 to $47,499',
    '$47,500 to $49,999'
]
brackets['$50,000 to $75,000'] = [
    '$50,000 to $52,499',
    '$52,500 to $54,999',
    '$55,000 to $57,499',
    '$57,500 to $59,999',
    '$60,000 to $62,499',
    '$62,500 to $64,999',
    '$65,000 to $67,499',
    '$67,500 to $69,999',
    '$70,000 to $72,499',
    '$72,500 to $74,999'
]
brackets['$75,000 to $100,000'] = [
    '$75,000 to $77,499',
    '$77,500 to $79,999',
    '$80,000 to $82,499',
    '$82,500 to $84,999',
    '$85,000 to $87,499',
    '$87,500 to $89,999',
    '$90,000 to $92,499',
    '$92,500 to $94,999',
    '$95,000 to $97,499',
    '$97,500 to $99,999'
]
brackets['$100,000 and over'] = [
    '$100,000 and over'
]

for bracket, cols in brackets.items():
    df[bracket] = df[cols].sum(axis=1)
    df[bracket + ' pct'] = df[bracket]/df['Totals']

pcts = df[['Race','Hispanic origin'] + [bracket + ' pct' for bracket in brackets.keys()]]

pcts
```

```python
pcts_filtered = pcts[(pcts['Hispanic origin'] == 'Non Hispanic') | (pcts['Race'] == 'Totals')]

pcts_filtered
```

```python
pcts_filtered.sum(axis=1)
```

```python
totals_filtered = df[(pcts['Hispanic origin'] == 'Non Hispanic') | (pcts['Race'] == 'Totals')]['Totals']

totals_filtered-pcts_filtered.sum(axis=1)*totals_filtered
```
